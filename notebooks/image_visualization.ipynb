# Visualizations

This notebook is a collection of image transformations and visualizations that can be used to understand the data better.

```python
import sys
from path_utils import add_parent_path_to_sys_path
# add parent directory to sys.path
current_path = sys.path[0]
add_parent_path_to_sys_path(current_path, verbose=False)

from utils.image_utils import convert_RGB_to_Grayscale

import matplotlib.pyplot as plt

from PIL import Image
```

```python
img_folder = '../../data/'
img_rgb_folder = img_folder + 'pokemon-rgb/'
img_gray_folder = img_folder + 'pokemon-grayscale/'
```

```python
img_rgb_path = img_folder + '/pokemon-rgb/1/1-0.png'

# Open the image file
image_rgb = Image.open(img_rgb_path)
# convert to grayscale
image_gray = image_rgb.convert('L')

# plot both images
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(4, 3))
axes[0].imshow(image_rgb)
axes[0].set_title('RGB')
axes[1].imshow(image_gray, cmap='gray')
axes[1].set_title('Grayscale')
plt.show()
image_rgb.size

convert_RGB_to_Grayscale(img_rgb_folder, img_gray_folder, verbose=False)
```
